```python
# %load script.py
# names of hurricanes
names = ['Cuba I', 'San Felipe II Okeechobee', 'Bahamas', 'Cuba II', 'CubaBrownsville', 'Tampico', 'Labor Day', 'New England', 'Carol', 'Janet', 'Carla', 'Hattie', 'Beulah', 'Camille', 'Edith', 'Anita', 'David', 'Allen', 'Gilbert', 'Hugo', 'Andrew', 'Mitch', 'Isabel', 'Ivan', 'Emily', 'Katrina', 'Rita', 'Wilma', 'Dean', 'Felix', 'Matthew', 'Irma', 'Maria', 'Michael']

# months of hurricanes
months = ['October', 'September', 'September', 'November', 'August', 'September', 'September', 'September', 'September', 'September', 'September', 'October', 'September', 'August', 'September', 'September', 'August', 'August', 'September', 'September', 'August', 'October', 'September', 'September', 'July', 'August', 'September', 'October', 'August', 'September', 'October', 'September', 'September', 'October']

# years of hurricanes
years = [1924, 1928, 1932, 1932, 1933, 1933, 1935, 1938, 1953, 1955, 1961, 1961, 1967, 1969, 1971, 1977, 1979, 1980, 1988, 1989, 1992, 1998, 2003, 2004, 2005, 2005, 2005, 2005, 2007, 2007, 2016, 2017, 2017, 2018]
unique_years = set(years)
print(unique_years)
print('\n')            
# maximum sustained winds (mph) of hurricanes
max_sustained_winds = [165, 160, 160, 175, 160, 160, 185, 160, 160, 175, 175, 160, 160, 175, 160, 175, 175, 190, 185, 160, 175, 180, 165, 165, 160, 175, 180, 185, 175, 175, 165, 180, 175, 160]

# areas affected by each hurricane
areas_affected = [['Central America', 'Mexico', 'Cuba', 'Florida', 'The Bahamas'], ['Lesser Antilles', 'The Bahamas', 'United States East Coast', 'Atlantic Canada'], ['The Bahamas', 'Northeastern United States'], ['Lesser Antilles', 'Jamaica', 'Cayman Islands', 'Cuba', 'The Bahamas', 'Bermuda'], ['The Bahamas', 'Cuba', 'Florida', 'Texas', 'Tamaulipas'], ['Jamaica', 'Yucatn Peninsula'], ['The Bahamas', 'Florida', 'Georgia', 'The Carolinas', 'Virginia'], ['Southeastern United States', 'Northeastern United States', 'Southwestern Quebec'], ['Bermuda', 'New England', 'Atlantic Canada'], ['Lesser Antilles', 'Central America'], ['Texas', 'Louisiana', 'Midwestern United States'], ['Central America'], ['The Caribbean', 'Mexico', 'Texas'], ['Cuba', 'United States Gulf Coast'], ['The Caribbean', 'Central America', 'Mexico', 'United States Gulf Coast'], ['Mexico'], ['The Caribbean', 'United States East coast'], ['The Caribbean', 'Yucatn Peninsula', 'Mexico', 'South Texas'], ['Jamaica', 'Venezuela', 'Central America', 'Hispaniola', 'Mexico'], ['The Caribbean', 'United States East Coast'], ['The Bahamas', 'Florida', 'United States Gulf Coast'], ['Central America', 'Yucatn Peninsula', 'South Florida'], ['Greater Antilles', 'Bahamas', 'Eastern United States', 'Ontario'], ['The Caribbean', 'Venezuela', 'United States Gulf Coast'], ['Windward Islands', 'Jamaica', 'Mexico', 'Texas'], ['Bahamas', 'United States Gulf Coast'], ['Cuba', 'United States Gulf Coast'], ['Greater Antilles', 'Central America', 'Florida'], ['The Caribbean', 'Central America'], ['Nicaragua', 'Honduras'], ['Antilles', 'Venezuela', 'Colombia', 'United States East Coast', 'Atlantic Canada'], ['Cape Verde', 'The Caribbean', 'British Virgin Islands', 'U.S. Virgin Islands', 'Cuba', 'Florida'], ['Lesser Antilles', 'Virgin Islands', 'Puerto Rico', 'Dominican Republic', 'Turks and Caicos Islands'], ['Central America', 'United States Gulf Coast (especially Florida Panhandle)']]

# damages (USD($)) of hurricanes
damages = ['Damages not recorded', '100M', 'Damages not recorded', '40M', '27.9M', '5M', 'Damages not recorded', '306M', '2M', '65.8M', '326M', '60.3M', '208M', '1.42B', '25.4M', 'Damages not recorded', '1.54B', '1.24B', '7.1B', '10B', '26.5B', '6.2B', '5.37B', '23.3B', '1.01B', '125B', '12B', '29.4B', '1.76B', '720M', '15.1B', '64.8B', '91.6B', '25.1B']

# deaths for each hurricane
deaths = [90,4000,16,3103,179,184,408,682,5,1023,43,319,688,259,37,11,2068,269,318,107,65,19325,51,124,17,1836,125,87,45,133,603,138,3057,74]

# write your update damages function here:
def convert_damages(damages):
    updated_damages = []
    for value in damages:
        if value == 'Damages not recorded':
            updated_damages.append('Damages not recorded')
        elif 'M' in value:
            value = float(value.strip('M')) * 10**6
            updated_damages.append(value)
        elif 'B' in value:
            value = float(value.strip('B')) * 10**9
            updated_damages.append(value)
        else:
            value = float(value)
            updated_damages.append(value)
    return updated_damages
updated_damages = convert_damages(damages)
print(updated_damages)
print('\n')            
# write your construct hurricane dictionary function here:
hurricane_dict = {}
for i in range(0, len(names)):
    hurricane_dict.update({names[i]: {'Month': months[i], 'Year': years[i], 'Max Sustained Wind':max_sustained_winds[i], 'Areas Affected':areas_affected[i], 'Damage': updated_damages[i], 'Death':deaths[i]}})
print(hurricane_dict)
print('\n')            
hurricanes = []
for name in hurricane_dict:
    entry = {'Name': name}
    entry.update(hurricane_dict.get(name))
    hurricanes.append(entry)
print(hurricanes)
print('\n')            
# write your construct hurricane by year dictionary function here:
def hurricanes_by_year(hurricane_dict):
    hurricanes = []
    for name in hurricane_dict:
        entry = {'Name': name}
        entry.update(hurricane_dict.get(name))
        hurricanes.append(entry)
    hurricanes_by_year = {}
    for year in unique_years:
        hurricanes_that_year = []
        for hurricane in hurricanes:
            if hurricane.get('Year') == year:
                hurricanes_that_year.append(hurricane)
        hurricanes_by_year.update({year: hurricanes_that_year})
    return hurricanes_by_year
hurricanes_by_year = hurricanes_by_year(hurricane_dict)
print(hurricanes_by_year)
print('\n')            
# write your count affected areas function here:
def count_affected_areas(hurricane_dict):
    areas = []
    for name in hurricane_dict:
        areas += hurricane_dict[name]['Areas Affected']
    del areas[-1]
    unique_areas = set(areas)
    hit_count = {}
    for area in unique_areas:
        hit_count.update({area: areas.count(area)})
    return hit_count
affected_areas_count = count_affected_areas(hurricane_dict)
print('\n')
print(affected_areas_count)
# write your find most affected area function here:
def most_affected(affected_areas_count):
    most_affected = {}
    num_hits = 0
    for area, count in affected_areas_count.items():
        if count == num_hits:
            most_affected.update({area: count})
        elif count > num_hits:
            most_affected.clear()
            num_hits = count
            most_affected.update({area: count})
        else:
            continue
    return most_affected
most_affected = most_affected(affected_areas_count)
print('\n')
print(most_affected)
# write your greatest number of deaths function here:
def most_deadly(hurricane_dict):
    deaths = {}
    for name, value in hurricane_dict.items():
        deaths.update({name: value.get('Death')})
    most_deadly = {}
    death_count = 0
    for name, death in deaths.items():
        if death == death_count:
            most_deadly.update({name: death})
        elif death > death_count:
            most_deadly.clear()
            death_count = death
            most_deadly.update({name: death})
        else:
            continue
    return most_deadly
most_deadly = most_deadly(hurricane_dict)
print('\n')
print(most_deadly)
# write your catgeorize by mortality function here:
def mortality_rating(hurricane_dict):
    hurricanes = []
    for name in hurricane_dict:
        entry = {'Name': name}
        entry.update(hurricane_dict.get(name))
        hurricanes.append(entry)
    hurricanes_by_mortality = {}
    rating0 = []
    rating1 = []
    rating2 = []
    rating3 = []
    rating4 = []
    rating5 = []
    for hurricane in hurricanes:
        if hurricane.get('Death') > 10000:
            rating5.append(hurricane)
            hurricanes_by_mortality.update({5: rating5})
        elif hurricane.get('Death') > 1000:
            rating4.append(hurricane)
            hurricanes_by_mortality.update({4: rating4})
        elif hurricane.get('Death') > 500:
            rating3.append(hurricane)            
            hurricanes_by_mortality.update({3: rating3})
        elif hurricane.get('Death') > 100:
            rating2.append(hurricane)            
            hurricanes_by_mortality.update({2: rating2})
        elif hurricane.get('Death') > 0:
            rating1.append(hurricane)            
            hurricanes_by_mortality.update({1: rating1})
        else:
            rating0.append(hurricane)            
            hurricanes_by_mortality.update({0: rating0})
    return hurricanes_by_mortality
hurricanes_by_mortality_rating = mortality_rating(hurricane_dict)
print('\n')
print(hurricanes_by_mortality_rating)
# write your greatest damage function here:
def most_damaging(hurricane_dict):
    damages = {}
    for name, value in hurricane_dict.items():
        damages.update({name: value.get('Damage')})
    most_damaging = {}
    damage_cost = 0
    for name, damage in damages.items():
        if damage == 'Damages not recorded':
            continue
        if damage == damage_cost:
            most_damaging.update({name: damage})
        elif damage > damage_cost:
            most_damaging.clear()
            damage_cost = damage
            most_damaging.update({name: damage})
        else:
            continue
    return most_damaging
most_costly = most_damaging(hurricane_dict)
print('\n')
print(most_costly)
# write your catgeorize by damage function here:
damage_scale = {0: 0,
                1: 100000000,
                2: 1000000000,
                3: 10000000000,
                4: 50000000000}
def categorize_damage(hurricane_dict):
    hurricanes = []
    for name in hurricane_dict:
        entry = {'Name': name}
        entry.update(hurricane_dict.get(name))
        hurricanes.append(entry)
    hurricanes_by_damage = {}
    category0 = []
    category1 = []
    category2 = []
    category3 = []
    category4 = []
    category5 = []
    for hurricane in hurricanes:
        if hurricane.get('Damage') == 'Damages not recorded':
            category0.append(hurricane)
            hurricanes_by_damage.update({0: category0})
        elif hurricane.get('Damage') > 50000000000:
            category5.append(hurricane)
            hurricanes_by_damage.update({5: category5})
        elif hurricane.get('Damage') > 10000000000:
            category4.append(hurricane)
            hurricanes_by_damage.update({4: category4})
        elif hurricane.get('Damage') > 1000000000:
            category3.append(hurricane)            
            hurricanes_by_damage.update({3: category3})
        elif hurricane.get('Damage') > 100000000:
            category2.append(hurricane)            
            hurricanes_by_damage.update({2: category2})
        elif hurricane.get('Damage') > 0:
            category1.append(hurricane)            
            hurricanes_by_damage.update({1: category1})
        else:
            category0.append(hurricane)
            hurricanes_by_damage.update({0: category0})
    return hurricanes_by_damage
hurricanes_by_damage = categorize_damage(hurricane_dict)
print('\n')
print(hurricanes_by_damage)
```
